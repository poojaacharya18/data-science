```python
from os import ST_NOSUID
print('import the libraries')
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
sns.set(color_codes=True)
```

```python
print('load the dataset using pandas read_csv() function')
car=pd.read_csv('/content/drive/MyDrive/principle of data science lab/poo.csv')
car
```

```python
print('Display the head of the dataset using')
car.head()
```

```python
print('Display the bottom 5 row from the data set using the tail() function"')
car.tail(5)
```

```python
print("Print summary stattistics of the dataset using the describe() function")
car.describe(include='all')
```

```python
print('plot a histogram for all the variables')
car.hist(figsize=(20,30))
```

```python
print('box plot visulization the relation between vechile size and engine hp')
sns.boxplot(x='cyl',y='hp',data=car)
```

```python
sns.pairplot(car)
```

```python
print('Drop irrelvnt column from the dataset using drop() function')
car=car.drop(['mp','cycle','disp','hp','drat','wt','qsec','vs','am','gear','carb'],axis=1)
car.head(5)
```

```python
print('use rename() function to rename the column')
car=car.rename(columns={"mpg":"mp","cyl":"cycle"})
car.head(5)
```

```python
dup_row_car=car[car.duplicated()]
print('no.of duplicate row:',dup_row_car.shape)
```

```python
car.count()
```

```python
car=car.drop_duplicates()
car.head()
```

```python
print(car.isnull().sum())
```

```python
car=car.dropna()
car.count()
```

```python
car.disp.value_counts().nlargest(40).plot(kind='bar',figsize=(10,5))
plt.title("no.of carby disp")
plt.ylabel('no.of cars')
plt.xlabel('disp')
```
